What is the largest prime factor of the number 600851475143 ?

```python
def is_prime(x: int) -> bool:
    
    """Check if the number is prime."""
    
    for i in range(2, int(x**0.5)+1):
        if x % i == 0:
            return False
    return True
```

```python
def factorization(x: int) -> list:
    
    """Find all factors of number."""
    
    factors = []
    for i in range(1, int(x**0.5)+1):
        if x % i == 0:
            factors.append(i)
            factors.append(int(x/i))
    if int(x**0.5)**2 == x:
        factors.remove(int(x**0.5))
    return factors
```

# Problem 14: Longest Collatz sequence

The following iterative sequence is defined for the set of positive integers:

n → n/2 (n is even)
n → 3n + 1 (n is odd)

Using the rule above and starting with 13, we generate the following sequence:

13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1
It can be seen that this sequence (starting at 13 and finishing at 1) contains 10 terms. Although it has not been proved yet (Collatz Problem), it is thought that all starting numbers finish at 1.

Which starting number, under one million, produces the longest chain?

NOTE: Once the chain starts the terms are allowed to go above one million.

```python
def length_Collatz_sequence(n: int) -> int:
    length = 1
    while n != 1:
        if n % 2 == 0:
            n = int(n/2)
        else:
            n = 3*n + 1
        length += 1
    return length
```

Saving February alone,
Which has twenty-eight, rain or shine.
And on leap years, twenty-nine.
A leap year occurs on any year evenly divisible by 4, but not on a century unless it is divisible by 400.
How many Sundays fell on the first of the month during the twentieth century (1 Jan 1901 to 31 Dec 2000)?

```python
def is_leap_year(year: int) -> bool:
    return (year % 4 == 0) and ((year % 100 != 0) or (year % 400 == 0))
```

```python
def generate_year(year: int) -> list:
    year_dates = {}
    thirty_days_months = [4, 6, 9, 11]
    for month in range(1, 13):
        if month in thirty_days_months:
            month_dates = list(map(lambda day: '{}-{}-{}'.format(year, month, day), range(1, 31)))
            month_dates_dict = dict(zip(month_dates, [None] * len(month_dates)))
            year_dates.update(month_dates_dict)
        else:
            if month == 2:
                if is_leap_year(year):
                    month_dates = list(map(lambda day: '{}-{}-{}'.format(year, month, day), range(1, 30)))
                    month_dates_dict = dict(zip(month_dates, [None] * len(month_dates)))
                    year_dates.update(month_dates_dict)
                else:
                    month_dates = list(map(lambda day: '{}-{}-{}'.format(year, month, day), range(1, 29)))
                    month_dates_dict = dict(zip(month_dates, [None] * len(month_dates)))
                    year_dates.update(month_dates_dict)
            else:
                month_dates = list(map(lambda day: '{}-{}-{}'.format(year, month, day), range(1, 32)))
                month_dates_dict = dict(zip(month_dates, [None] * len(month_dates)))
                year_dates.update(month_dates_dict)
    return year_dates
```

Find the sum of all the numbers that can be written as the sum of fifth powers of their digits.

```python
def generate_digit_powers(shape: int) -> list:
    
    def define_supremum():
        n = 1
        while n <= len(str(n*9**shape)):
            n += 1
        return n
    
    digit_powers = []
    for n in range(2, 10**define_supremum()):
        sum_digit_powers = sum(map(lambda x: int(x)**shape, list(str(n))))
        if n == sum_digit_powers:
            digit_powers.append(sum_digit_powers)
    return digit_powers
```

# Problem 34: Digit factorials

145 is a curious number, as 1! + 4! + 5! = 1 + 24 + 120 = 145.

Find the sum of all numbers which are equal to the sum of the factorial of their digits.

Note: as 1! = 1 and 2! = 2 are not sums they are not included.

```python
def num_to_sum_factorials(n: int) -> int:
    return sum(map(lambda x: factorial(int(x)), list(str(n))))

def count_zeros_and_ones_in_number(n: int) -> int:
    return len(list(filter(lambda x: (x == '1') or (x == '0'), list(str(n)))))

for i in range(1, 1000001):
    if i == num_to_sum_factorials(i):
        print(i, num_to_sum_factorials(i))
```
